```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# -------------------------
# 🍎🍊 Simulate Fruit Data
# -------------------------

np.random.seed(42)

# Apples: heavier and redder
apple_mean = [160, 220]
apple_cov = [[25, 10], [10, 30]]
apples = np.random.multivariate_normal(apple_mean, apple_cov, 100)

# Oranges: lighter and less red
orange_mean = [130, 160]
orange_cov = [[25, -5], [-5, 20]]
oranges = np.random.multivariate_normal(orange_mean, orange_cov, 100)

# Combine all fruits into one dataset
fruits = np.vstack((apples, oranges))
weights = fruits[:, 0].reshape(-1, 1)  # Input feature: weight (X)
colors = fruits[:, 1]                  # Target value: color intensity (y)

# -------------------------
# 🔍 Learn the Linear Model
# -------------------------

# Fit a linear regression model: Color = θ₀ + θ₁ × Weight
model = LinearRegression()
model.fit(weights, colors)

# Predict colors from weights using the learned model
predicted_colors = model.predict(weights)

# -------------------------
# 📏 Evaluate the Model
# -------------------------

mse = mean_squared_error(colors, predicted_colors)
mae = mean_absolute_error(colors, predicted_colors)
r2 = r2_score(colors, predicted_colors)

# -------------------------
# 📈 Visualize the Results
# -------------------------

plt.figure(figsize=(10, 6))
plt.scatter(weights, colors, color="gray", alpha=0.6, label="Actual Fruits")
plt.plot(weights, predicted_colors, color="red", label="Regression Line")
plt.xlabel("Weight (grams)")
plt.ylabel("Color Intensity")
plt.title("Predicting Fruit Color from Weight")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------------
# 🧠 Print Summary
# -------------------------

print("Learned Model:")
print(f"  Color ≈ {model.intercept_:.2f} + {model.coef_[0]:.2f} × Weight")
print("\nPerformance Metrics:")
print(f"  Mean Squared Error (MSE): {mse:.2f}")
print(f"  Mean Absolute Error (MAE): {mae:.2f}")
print(f"  R-squared (R²): {r2:.3f}")
```
